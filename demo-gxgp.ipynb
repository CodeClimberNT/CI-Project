```python
import operator
import numpy as np

from icecream import ic
from gxgp import Node
```

```python
tree = Node(np.multiply, [Node(np.add, [Node(10), Node("x")]), Node(2)])
tree.draw()
```

```python
tree(x=np.array([1, 2, 3, 4, 5]))
```

```python
tree2 = Node(
    lambda a, b, c, d, e, f: ((a + b * c - d) / e) + f,
    [Node("x"), Node("y"), Node(3), Node("z"), Node(2), Node("foo")],
)
tree2.draw()
```

```python
# Explicit keyword arguments (unused arguments are ignored)
tree2(x=1, y=2, z=3, foo=42)
# (a + b * c - d) / e
#  1   2   3   3    2
```

```python
(1 + 2 * 3 - 3) / 2
```

```python
# Unpacking a dictionary
vars = {"x": 1, "y": 2, "z": 3, "foo": 42}
tree2(**vars)
```
